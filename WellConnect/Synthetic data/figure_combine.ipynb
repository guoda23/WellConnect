```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img1 = mpimg.imread('plots/GPT-4o/conditional_joint_probabilities_heatmap_GPT-4o.png')
img2 = mpimg.imread('plots/Gemini 2.5 Pro/conditional_joint_probabilities_heatmap_Gemini 2.5 Pro.png')

# Set figure dimensions based on image shape
dpi = 300
fig_width = max(img1.shape[1], img2.shape[1]) / dpi
fig_height = (img1.shape[0] + img2.shape[0]) / dpi + 0.5  # extra space for title

fig, axes = plt.subplots(nrows=2, figsize=(fig_width, fig_height), dpi=dpi)

# Display images
axes[0].imshow(img1)
axes[0].axis('off')
axes[1].imshow(img2)
axes[1].axis('off')

fig.suptitle("   P(Age Group × Sex × Origin | Depression): Benchmark vs Synthetic Datasets",
             fontsize=30, y=0.98, ha='center')
plt.subplots_adjust(top=0.94, hspace=0.02) # Adjust spacing between plots and title
plt.savefig('plots/conditional_joint_probabilities_combined.png', dpi=dpi, bbox_inches='tight', pad_inches=0.05)
plt.show()
```
