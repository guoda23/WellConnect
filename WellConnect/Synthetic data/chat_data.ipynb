```python
def show_percentage_distribution(column):
    print(f"=== {column} (percentage) ===")
    print(demographics[column].value_counts(normalize=True).sort_index() * 100, "\n")

# List of columns to inspect
columns_to_check = [
    "AgeGroup", "SexAtBirth", "GenderIdentity", "Ethnicity", "Education",
    "WorkStatus", "District", "Religion", "EmotionalStability"
]

# Print percentage distributions
for col in columns_to_check:
    show_percentage_distribution(col)
```

by severity

```python
df_week1 = weekly_responses[weekly_responses["Week"] == 1].copy()

# Identify PHQ-9 item columns
phq_columns = [col for col in df_week1.columns if col.startswith("PHQ_")]

# Calculate total PHQ-9 score per participant
df_week1["PHQ9_Total"] = df_week1[phq_columns].sum(axis=1)

# Define a function to assign severity categories
def phq_category(score):
    if score <= 4:
        return "None"
    elif score <= 9:
        return "Mild"
    elif score <= 14:
        return "Moderate"
    elif score <= 19:
        return "Moderately Severe"
    else:
        return "Severe"

# Apply category function
df_week1["PHQ9_Category"] = df_week1["PHQ9_Total"].apply(phq_category)

# Get category distribution
severity_counts = df_week1["PHQ9_Category"].value_counts().sort_index()
print(severity_counts)

# Optional: Save summary to CSV
severity_counts.to_csv("phq9_severity_summary.csv", header=True)
```
